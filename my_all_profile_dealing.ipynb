```python
def str_to_dict(headers):
    ##将headers 的字符串转换城字典参数形式
    headers = headers.split("\n")
    d_headers = {}
    for h in headers:
        h = h.strip()
        if h:
            k,v = h.split(":",1)
            d_headers[k] = v.strip()
    return d_headers
```
